```python
import json # reading in source file
import pandas as pd
```

```python
Movie_Dir = "/kaggle/input/imdb-review-dataset/" 
json_files = ["part-01.json", "part-02.json", "part-03.json", "part-04.json", "part-05.json", "part-06.json"]
movie_list = list()

for j_file in json_files:
    with open(f"{Movie_Dir}/{j_file}", mode='r') as file:
        new_reviews = json.load(file)
        for review in new_reviews:
            movie_list.append(review)
            
# Convert list of dicts to pandas dataframe
df = pd.DataFrame(movie_list)
```

```python
df.head(1)
```

1. Which movie had the highest rating?

```python
df['rating'] = pd.to_numeric(df['rating'])
H_rating = df.sort_values('rating',ascending=False)
data = H_rating[["movie","rating","review_detail"]]
data.head()
```

```python
group = data.groupby(['movie', 'rating']).size().reset_index(name='count')
high_rating = group.sort_values(by=['rating','count'],ascending=False)
high_rating
```

```python
hr = high_rating['movie'].head(1)
print('Highest Rated Movie:',hr.to_string(index=False))
```

2. How many reviewers gave it the highest rating?

```python
HR = data.loc[df['movie'] == 'Dil Bechara (2020)'] 
print(HR['rating'].value_counts()[10])
```

3. What’s its least rating and review detail?

```python
LR = HR.sort_values("rating",ascending=True).head(1)

lr = LR['rating']
rd = LR['review_detail']
print(('rating:'),lr.to_string(index=False))
print(('review_detail:'),rd.to_string(index=False))
```
